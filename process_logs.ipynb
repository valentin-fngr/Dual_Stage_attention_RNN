```python
import tensorboard as tb
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
def plot_test_RMSE_over_time(paths, t_list): 

    y = []
    x = []
    
    for path, t in zip(paths, t_list): 
        df = pd.read_csv(path)
        rmse = df["Value"].values
        y.append(rmse) 
        x.append(t)

    plt.plot(x, y)
    plt.xticks(ticks=x, labels=[str(val) for val in x])
    plt.scatter(x, y)
    plt.xlabel('timesteps')
    plt.ylabel('RMSE')

    plt.title("RMSE w.r.t timesteps")


def plot_test_MAE_over_time(paths, t_list): 

    y = []
    x = []
    
    for path, t in zip(paths, t_list): 
        df = pd.read_csv(path)
        mae = df["Value"].values
        y.append(mae) 
        x.append(t)

    fig = plt.figure()

    plt.plot(x, y)
    plt.xticks(ticks=x, labels=[str(val) for val in x])
    plt.scatter(x, y)
    plt.xlabel('timesteps')
    plt.ylabel('MAE')

    
    plt.title("MAE w.r.t timesteps")
    plt.show()



def plot_test_MAPE_over_time(paths, t_list): 

    y = []
    x = []
    
    for path, t in zip(paths, t_list): 
        df = pd.read_csv(path)
        mape = df["Value"].values
        y.append(mape) 
        x.append(t)

    fig = plt.figure()

    plt.plot(x, y)
    plt.xticks(ticks=x, labels=[str(val) for val in x])
    plt.scatter(x, y)
    plt.xlabel('timesteps')
    plt.ylabel('MAPE')

    
    plt.title("MAPE w.r.t timesteps")
    plt.show()



def plt_test_prediction(path_pred, t_list, gt_path): 

    plt.figure(figsize=(10, 6))

    for path, t in zip(path_pred, t_list): 
        df = pd.read_csv(path)
        plt.plot(df["Step"], df["Value"], label=f"supervised_t={t}")

    df_true = pd.read_csv(gt_path) 
    plt.plot(df_true["Step"], df_true["Value"], label=f"ground truth", c="black")

    plt.ylabel('Y')
    plt.xlabel('Step')
    plt.ylabel('Value')
    plt.title('Test Predictions vs Steps')
    plt.legend()
    plt.show()

```
